```python
import pandas as pd
import faker
import random
from datetime import datetime, timedelta
import re
```

```python
# Initialize Faker
fake = faker.Faker()
```

```python
# Define categories
categories = ['Business', 'Transaction & Security', 'Order Confirmation', 'Subscription Renewals', 'Promotional']
```

```python
# Define products for promotional emails
products = ['Smartphone', 'Laptop', 'Headphones', 'TV', 'Charger', 'Tablet']
```

```python
# Function to clean text
def clean_text(text):
    if isinstance(text, str):  # Check if the value is a string
        # Convert to lowercase
        text = text.lower()
        # Remove special characters and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        # Remove extra whitespace
        text = ' '.join(text.split())
    return text
```

```python
# Generate synthetic data
data = []
for _ in range(1000):  # Generate 1000 rows
    category = random.choice(categories)
    from_email = fake.email()
    subject = fake.sentence(nb_words=5)
    # Generate raw message (with special characters, extra spaces, etc.)
    raw_message = fake.paragraph(nb_sentences=3)
    raw_message += f"\n\nRegards,\n{fake.name()}"
    raw_message += f"\nPhone: {fake.phone_number()}"
    raw_message += f"\nWebsite: {fake.url()}"
    # Generate cleaned message
    cleaned_message = clean_text(raw_message)

    last_received_date = fake.date_time_between(start_date='-1y', end_date='now').strftime('%Y-%m-%d %H:%M:%S')
    action_awaited = None
    coupon = None
    product = None

    if category == 'Business':
        action_awaited = (datetime.strptime(last_received_date, '%Y-%m-%d %H:%M:%S') + timedelta(days=5)).strftime('%Y-%m-%d')
    elif category == 'Subscription Renewals':
        action_awaited = (datetime.strptime(last_received_date, '%Y-%m-%d %H:%M:%S') + timedelta(days=7)).strftime('%Y-%m-%d')
    elif category == 'Promotional':
        coupon = fake.bothify(text='???##').upper()
        product = random.choice(products)

    data.append({
        'Category': category,
        'From': from_email,
        'Subject': subject,
        'Raw Message': raw_message,
        'Cleaned Message': cleaned_message,
        'Last Received Date': last_received_date,
        'Action Awaited': action_awaited,
        'Coupon': coupon,
        'Products': product
    })
```

```python
df = pd.DataFrame(data)
```

```python
# Save to CSV
df.to_csv('/content/email_dataset.csv', index=False)
```
